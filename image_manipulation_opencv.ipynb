Resizing Images

The function resize resizes the image src down to or up to the specified size. The size and type are derived from the src,dsize,fx, and fy.
Function Syntax

dst = resize( src, dsize[, dst[, fx[, fy[, interpolation]]]] )

dst: output image; it has the size dsize (when it is non-zero) or the size computed from src.size(), fx, and fy; the type of dst is the same as of src.

The function has 2 required arguments:

    src: input image
    dsize: output image size

Optional arguments that are often used include:

    fx: Scale factor along the horizontal axis; when it equals 0, it is computed as ()./.
    fy: Scale factor along the vertical axis; when it equals 0, it is computed as ()./.

The output image has the size dsize (when it is non-zero) or the size computed from src.size(), fx, and fy; the type of dst is the same as of src.

there are 3 types of resizing :

    1. resize using fx anf fy to specify scaling factor
    2. specifying the exact size of the image
    3. resize with the same or specific aspect ratio

```python
# 1. resize using fx anf fy to specify scaling factor
method1 = cv2.resize(cropped_boat,None,fx = 2,fy = 2)
plt.imshow(method1)
```

```python
# specifying the exact size of the image
method2 = cv2.resize(cropped_boat,dsize = (200,300),interpolation = cv2.INTER_AREA)
plt.imshow(method2)
```

```python
# resize with the same or specific aspect ratio
desired_width = 100
aspect_ratio = desired_width / cropped_boat.shape[1]
desired_height = int(cropped_boat.shape[0] * aspect_ratio)
dim = (desired_width, desired_height)

method3 = cv2.resize(cropped_boat, dsize=dim, interpolation=cv2.INTER_AREA)
plt.imshow(method3)
```

```python
method3 =  method3[:,:,::-1]
cv2.imwrite("cropped_boat.jpg",method3)
Image(filename = "cropped_boat.jpg")
```

Flipping Images

The function flip flips the array in one of three different ways (row and column indices are 0-based):
Function Syntax

dst = cv.flip( src, flipCode )

dst: output array of the same size and type as src.

The function has 2 required arguments:

    src: input image
    flipCode: a flag to specify how to flip the array; 0 means flipping around the x-axis and positive value (for example, 1) means flipping around y-axis. Negative value (for example, -1) means flipping around both axes.


```python
# flipping the images
hori_flip = cv2.flip(boat_rgb,1)
verti_flip = cv2.flip(boat_rgb,0)
both_flip = cv2.flip(boat_rgb,-1)

plt.figure(figsize = [20,5])
plt.subplot(141);plt.imshow(hori_flip);plt.title("horizontal")
plt.subplot(142);plt.imshow(verti_flip);plt.title("vetical")
plt.subplot(143);plt.imshow(both_flip);plt.title("both")
plt.subplot(144);plt.imshow(boat_rgb);plt.title("original")
```
